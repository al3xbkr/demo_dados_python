# http://dados.gov.br

```python
import requests, zipfile, StringIO, pandas
```

```python
r = requests.get('http://www.justica.gov.br/dados-abertos/anexos/2015-01.zip')
```

```python
data = None
with zipfile.ZipFile(StringIO.StringIO(r.content)) as z:
    f = z.open('2015-01.csv')
    data = pd.read_table(f, sep=";", encoding="iso8859", decimal=',', usecols=['Cidade', 'Sexo', 'Nota do Consumidor'])
```

```python
data[data['Cidade'] == 'Porto Alegre'].groupby('Sexo').Sexo.count().plot(kind='bar')
```

## Eliminando os registros que não tem nota

```python
data = data[ data['Nota do Consumidor'].isnull() == False ]
```

## Projetando os dados num espaço reduzido

```python
data = data[['Nota do Consumidor','Cidade']]
```

## Filtrando somente as cidades que tem mais de 100 registros de nota

```python
df = data.groupby('Cidade').filter(lambda x: len(x) > 100).groupby('Cidade')
```

```python
df.mean()
```

```python
df.mean().plot(kind='bar')
```
